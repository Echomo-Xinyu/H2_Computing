```python
#Task 2.1
handle = open("TESTDATA.txt", "r")
data = []
for line in handle:
    data.append(line.split(","))
handle.close()

# print(len(data[0]))

def qs(dat):
    if len(dat) == 0:
        return []
    pivot = dat[0]
    less, more = [], []
    for i in range(1, len(dat)-1):
        if dat[i][0] >= pivot[0]:
            more.append(dat[i])
        else:
            less.append(dat[i])
    return qs(less) + [pivot] + qs(more)

sorted_data = qs(data)

handle = open("TESTDATA_Filename_order_SUN_XINYU_G1647712K-1703006.txt", "w")
for i in range(len(sorted_data)):
    handle.write(sorted_data[i][0] + ", " + data[i][1] + ", " + data[i][2] + "\n")
handle.close()
```

```python
#Task 2.2
# sort based on chronological
def qs_ch(dat):
    if len(dat) == 0:
        return []
    pivot = dat[0]
    less, more = [], []
    for i in range(1, len(dat)-1):
        if dat[i][1] >= pivot[1]:
            more.append(dat[i])
        else:
            less.append(dat[i])
    return qs_ch(less) + [pivot] + qs_ch(more)

sorted_data_ch = qs_ch(data)
for i in range(len(sorted_data_ch)):
    print(sorted_data_ch[i][0] + ", " + sorted_data_ch[i][1] + ", " + sorted_data_ch[i][2])

while True:
    input_name = input("Please type the file you want to find:\n")
    whether_found = False
    for i in range(len(data)):
        if data[i][0] == input_name:
            print(data[i][1] + " " + data[i][2])
            whether_found = True
    if whether_found:
        break
    print("Fail to find the file with such name. Please try again")


input_date = input("Please input the data last modified you want to find, in the format of YYYY-MM-DD:\n")
for i in range(len(data)):
    if data[i][1] == input_date:
        print(data[i][0] + " " + data[i][2])
    
    
```

```python
#Task 2.3
# insertion sort
def insertionSort(L):
    n = len(L)-1
    for i in range(1, n+1):
        for j in range(i, 0, -1):
            if L[j] < L[j-1]:
                L[j], L[j-1] = L[j-1], L[j]
    return L
```
